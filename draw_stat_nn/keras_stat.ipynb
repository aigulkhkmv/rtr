```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import math
import numpy as np
import pandas
```

```python
with open("/home/aigul/Retro/stat_from_keras/80_200_325_1*2000/accuracy.txt", "r") as ac:
    accur = ac.read()
dpi = 80
fig = plt.figure(dpi = dpi, figsize = (512 / dpi, 384 / dpi) )
plt.title('Accuracy_80_2000_325_1*2000')
mpl.rcParams.update({'font.size': 15})
ax = plt.axes()
ax.yaxis.grid(True, zorder = 1)
plt.bar([i for i in range(80)], [float(n) for n in accur.split("\n")[:len(accur.split("\n"))-1]],  color = 'green', alpha = 1,
        zorder = 2)
plt.savefig('/home/aigul/Retro/stat_from_keras/80_200_325_1*2000/accuracy_80_200_325_1*2000.png')
```
